### Numerical methods for solving  ODEs

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Euler's method

Euler's method approximates the solution of the ODE 
$$
\frac{dy}{dt} = f(t, y)
$$
with the formula:
$$
y_{n+1} = y_n + h \cdot f(t_n, y_n)
$$


```python
def f(t, y):
    return -y
t0, y0 = 0, 1  # Initial conditions
t_end = 5       # End time
h = 0.1         # Step size
t_values = np.arange(t0, t_end + h, h)
y_values = np.zeros(len(t_values))
y_values[0] = y0
# euler's method
for i in range(1, len(t_values)):
    y_values[i] = y_values[i-1] + h * f(t_values[i-1], y_values[i-1])
#analytical method
y_analytical = y0 * np.exp(-t_values)
```

```python
plt.plot(t_values, y_values, label="Euler's Method")
plt.plot(t_values, y_analytical, label="Analytical Solution", linestyle='dashed')
plt.xlabel('Time')
plt.ylabel('y(t)')
plt.legend()
plt.title("Euler's Method for dy/dt = -y")
plt.show()
```

#### Runge-Kutta Method (RK4)
The 4th-order Runge-Kutta (RK4) method is more accurate than Euler's method. It uses intermediate steps to compute the solution:
$$
k_1 = h \cdot f(t_n, y_n)
$$

$$
k_2 = h \cdot f\left(t_n + \frac{h}{2}, y_n + \frac{k_1}{2}\right)
$$

$$
k_3 = h \cdot f\left(t_n + \frac{h}{2}, y_n + \frac{k_2}{2}\right)
$$

$$
k_4 = h \cdot f\left(t_n + h, y_n + k_3\right)
$$

$$
y_{n+1} = y_n + \frac{k_1 + 2k_2 + 2k_3 + k_4}{6}
$$

```python
def f(t, y):
    return -y
t0, y0 = 0, 1  
t_end = 5       # End time
h = 0.1         # Step size
t_values = np.arange(t0, t_end + h, h)
y_values = np.zeros(len(t_values))
y_values[0] = y0

# RK4 method
for i in range(1, len(t_values)):
    t_n, y_n = t_values[i-1], y_values[i-1]
    k1 = h * f(t_n, y_n)
    k2 = h * f(t_n + h/2, y_n + k1/2)
    k3 = h * f(t_n + h/2, y_n + k2/2)
    k4 = h * f(t_n + h, y_n + k3)
    y_values[i] = y_n + (k1 + 2*k2 + 2*k3 + k4) / 6

# Analytical solution 
y_analytical = y0 * np.exp(-t_values)
```

```python
plt.plot(t_values, y_values, label="RK4 Method")
plt.plot(t_values, y_analytical, label="Analytical Solution", linestyle='dashed')
plt.xlabel('Time')
plt.ylabel('y(t)')
plt.legend()
plt.title("Runge-Kutta (RK4) Method for dy/dt = -y")
plt.show()
```
